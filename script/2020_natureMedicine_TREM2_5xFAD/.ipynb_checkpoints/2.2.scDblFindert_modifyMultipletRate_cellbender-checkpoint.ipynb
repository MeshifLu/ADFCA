# Add cellN

```python
### remove ambient RNAs, basic CellQC (min_genes= 200, min_Cell=3), remove doublets using decontx dataset

remainCellN_L = []
for index, row in decontxSummaryDf.iterrows():
    expName=row[0]
    # matrixName = f'{expName}_filtered_feature_bc_matrix'
    print(expName,)
    
    ### Use decontX data
    decontxMtx=f'{allDecontxFolder}/{expName}'
    # print(decontxMtx)
    adataDecontx = sc.read_10x_mtx(decontxMtx, var_names='gene_symbols', cache=True)
    adataDecontx.var_names_make_unique()
    # output cellN
    # basic QC
    sc.pp.filter_cells(adataDecontx, min_genes=200)
    sc.pp.filter_genes(adataDecontx, min_cells=3)
    
    ### remove doublets
    print(len(adataDecontx))
    metadataOutPath=f'{metadataFolder}/{expName}.csv'
    metadataDf = pd.read_csv(metadataOutPath, index_col=0)
    
    keep_L = metadataDf[ metadataDf['scDblFinder.class'] == 'singlet' ].index.to_list()
    adataDecontx = adataDecontx[adataDecontx.obs_names.isin(keep_L)]
    
    # Modify nan gene name
    if True in list(adataDecontx.var.index.isna()):
        print('nan found in index!!!')
        nanNum = list(adataDecontx.var.index.isna()).index(True)
        adataDecontx.var.gene_ids = ['nan' if n == nanNum else var for n, var in enumerate(adataDecontx.var.gene_ids)]
        adataDecontx.var_names = ['nan' if n == nanNum else var for n, var in enumerate(adataDecontx.var_names)]
        print(len(adataDecontx))

    remainCellN_L.append(len(adataDecontx))
    adataDecontx.write(f'{scDblFinderFolder}/adata_decontx1_scDblFinder_{expName}.h5ad')
    
decontxSummaryDf['decontx1_scDblFinderCellN'] = remainCellN_L
```
